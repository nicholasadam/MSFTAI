```python
import time 
import requests
from __future__ import print_function
from PIL import Image
from io import BytesIO
# Import library to display results
import matplotlib.pyplot as plt
%matplotlib inline 
# Display images within Jupyter
```

```python
# Variables
_region = 'westcentralus' #Here you enter the region of your subscription
_url = 'https://{}.api.cognitive.microsoft.com/vision/v2.0/generateThumbnail'.format(_region)
_key = '9438c593902948179871edf1cf0033c6' #Here you have to paste your primary key
headers = dict()
headers['Ocp-Apim-Subscription-Key'] = _key
headers['Content-Type'] = 'application/octet-stream'
params = { 'width': '500','height': '500','smartCropping': 'true'} 
_maxNumRetries = 10
json=None
pathToFileInDisk = r'4.jpg'
with open( pathToFileInDisk, 'rb' ) as f:
    data = f.read()
```

```python
response = requests.request( 'post', _url, json = json, data = data, headers = headers, params = params )
```

```python
thumbnail = Image.open(BytesIO(data))

# Display the thumbnail.
plt.imshow(thumbnail)
plt.axis("off")

# Verify the thumbnail size.
print("Thumbnail is {0}-by-{1}".format(*thumbnail.size))
```

```python
thumbnail = Image.open(BytesIO(response.content))

# Display the thumbnail.
plt.imshow(thumbnail)
plt.axis("off")

# Verify the origin image size.
print("Origin Image is {0}-by-{1}".format(*thumbnail.size))
```
